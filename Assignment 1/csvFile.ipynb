```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,1].values
```

```python
X_train, X_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.2)
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
prediction = regressor.predict(X_test)
df = pd.DataFrame({"Actual Values": y_test, "Predicted Values": prediction})
```

```python
plt.scatter(X_train, Y_train, color = 'red', label = 'Real Salary')
plt.scatter(X_test, prediction, color = 'green', label = 'Predicted Salary')
plt.plot(X_train, regressor.predict(X_train), color = 'blue', label = 'Regression Model')
plt.title('Salary vs Year')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print(r2)
```
